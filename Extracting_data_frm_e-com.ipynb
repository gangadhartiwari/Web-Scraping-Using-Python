```python
import requests
from bs4 import BeautifulSoup
```

```python
# this is the url of the site 

url = "https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops"  
```

```python
# using get function to store the data into a variabe called data

data = requests.get(url)
data
```

```python
# using beBeautifulSoup for extractiong the data into structured format

soup = BeautifulSoup(data.text, "lxml")
```

```python
# print(soup)
```

```python
# Extracting the Names of the products (e.g. : Laptop)

Names = soup.find_all("a", class_ = "title")
```

```python
# Printing the itmes from the  Names list 

for i in Names: 
    print(i.text)
```

```python
# Extracting the prices of items 

Price = soup.find_all("h4", class_ = "price float-end card-title pull-right")
```

```python
for i in Price: 
    print(i.text)
```

```python
# Extracting the description of items

desc = soup.find_all("p", class_ = "description card-text")
```

```python
for i in desc: 
    print(i.text)
```
